```python
import vega_datasets
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn import linear_model
import sklearn.metrics as sm
```

```python
cr = vega_datasets.data.cars()
```

#Mengidendifikasi Data yang Rileate Satu Sama Lain

```python
sn.pairplot(cr)
```

#Cek Bagian "Horsepower" dan "Miles_per_Gallon" Apakah Ada Data yang Kosong

```python
set(cr['Horsepower'].isna())
```

```python
set(cr['Miles_per_Gallon'].isna())
```

#Mengisi Data yang Kosong di "Horsepower" dengan "Miles_per_Gallon" dengan Data Mean

```python
cr ['Miles_per_Gallon'] = cr['Miles_per_Gallon'].fillna(cr['Miles_per_Gallon'].mean())
cr ['Horsepower'] = cr['Horsepower'].fillna(cr['Horsepower'].mean())
```

#Membuat Model Regresi Linear

```python
lin_reg = linear_model.LinearRegression()
```

#Masuk Kedalam "sklern"

```python
lin_reg.fit(cr[['Displacement','Horsepower','Weight_in_lbs']],cr['Miles_per_Gallon'])
```

#Melihat Nilai "Displacement" dan "Horsepower"

```python
cr['Displacement'].describe()
```

```python
cr['Horsepower'].describe()
```

#Memprediksi Nilai "Miles_per_Galon" dari Variabel 'Displacement','Horsepower','Weight_in_lbs'

```python
lin_reg.predict([[100,200,4000]])
```

#Perhitungan Secara Matematis "y = a * x + b * x + c * x + int"

#Menentukan Nilai Coefisien dari Masing-masing "a,b,c"

```python
lin_reg.coef_
```
